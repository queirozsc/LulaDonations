## Webscrapping donors to Lula

```python
import requests
import json

url = 'https://doe.lula.com.br/api/campaign'

p = requests.get(url)
d = json.loads(p.content)

for doador in d['donors']:
    print(doador['nome'], doador['cpf'], doador['valor'], doador['data_envio'])
```

## Extracting information on the columns

```python
columns = 'cpf data_envio doador_id nome pagamento_tipo_id valor'.split()
columns
```

## Reading the data into Pandas

```python
import pandas as pd
donations = pd.DataFrame(d['donors'], columns=columns)
```

## Converting columns

```python
import numpy as np

def parse_float(x):
    try:
        x = float(x)
    except Exception:
        x = 0
    return x

donations['valor'] = donations['valor'].apply(parse_float)
```

```python
donations['date'] = pd.to_datetime(donations['data_envio'], infer_datetime_format=True)
```

```python
donations['week_start'] = donations['date'].dt.to_period('W')
donations.groupby('week_start')['valor'].sum()
```

## CPF

```python
dict_cpf = {
    0: 'RS',
    1: 'DF, GO, MS, TO',
    2: 'PA, AM, AC, AP, RO, RR',
    3: 'CE, MA, PI',
    4: 'PE, RN, PB, AL',
    5: 'BA, SE',
    6: 'MG',
    7: 'RJ, ES',
    8: 'SP',
    9: 'PR, SC'
}

def parse_cpf(cpf):
    try:
        source_district = dict_cpf[int(cpf[-3])]
    except Exception:
        source_district = 'UNDEFINED'
    return source_district

donations['district'] = donations['cpf'].apply(parse_cpf)
```

```python
donations.district.unique()
```

```python
donations_by_district = (donations.groupby(['district'])
    .size()
    .to_frame('donations')
    .reset_index()
    .sort_values(['district'])
)
donations_by_district
```

```python
values_by_district = (donations.groupby(['district'])['valor']
    .sum()
    .to_frame('R$')
    .reset_index()
    .sort_values(['R$'], ascending=[False]))
values_by_district
```

```python
donations_by_district = donations.groupby('district').agg({'valor': np.sum})
donations_by_district.sort_values('valor', ascending=False)
donations_by_district
```

## Ploting with Folium

```python
#!pip install folium
```

```python
import folium

brazil = folium.Map(
    location = [-16.1237611, -59.9219642],
    zoom_start=4
)

for name, donation in values_by_district.iterrows():
    if donation['district'] != 'UNDEFINED':
        folium.Marker(
            location = brazil_districts_location[donation['district'].split(',')[0]],
            icon = folium.Icon(color = colors[donation['district']]),
            popup = '<b>{}:</b> R$ {:.2f}</n>Doações: {}'.format(donation['district'], donation['R$'], donations_by_district[donation['district']]['donations'])
        ).add_to(brazil)
    
brazil
```

```python
folium.Marker(
    location=brazil_districts_location['MA'],
    icon=folium.Icon(color='red'),
    popup='<b>Maranhão</b>Doadores:</n>'
).add_to(brazil)

folium.Marker(
    location=brazil_districts_location['SC'],
    icon=folium.Icon(color='green'),
    popup='Santa Catarina'
).add_to(brazil)

brazil
```

```python
from folium import plugins

brazil_heatmap = folium.Map(
    location = brazil_districts_location['MA'],
    zoom_start = 11
)

brazil_heatmap.add_children(
    plugins.HeatMap(brazil_districts_location)
)

brazil_heatmap
```
